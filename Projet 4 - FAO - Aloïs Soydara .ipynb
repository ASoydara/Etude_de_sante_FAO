```python
#Importation de la libraire Pandas et des fichiers de données
import pandas as pd
pd.set_option("max_rows", None)
aide_ali = pd.read_csv('C:/Users/alois/Desktop/Formation Aloïs/Projet 4 - FAO/aide_alimentaire.csv')
dispo_ali = pd.read_csv('C:/Users/alois/Desktop/Formation Aloïs/Projet 4 - FAO/dispo_alimentaire.csv')
population = pd.read_csv('C:/Users/alois/Desktop/Formation Aloïs/Projet 4 - FAO/population.csv')
sous_nutri = pd.read_csv('C:/Users/alois/Desktop/Formation Aloïs/Projet 4 - FAO/sous_nutrition.csv')
```

```python
#Requete 1 : Proportion de personnes en état de sous nutrition

#Isolation de la population de l'année 2017
population_valeur = population['Année']==2017
population2017 = population[population_valeur]

#Isolation des valeurs de sous nutrition pour l'année 2017
sous_nutri_valeur = sous_nutri['Année']=='2016-2018'
sous_nutri2017 = sous_nutri[sous_nutri_valeur]

#Jointure des deux dataframes
jointure_pop_sn2017 = pd.merge(population2017, sous_nutri2017, on="Zone")

#Remplacement des '.' par des ',' et changement de nature de 'valeur_y' en numeric
jointure_pop_sn2017['Valeur_y']= jointure_pop_sn2017['Valeur_y'].str.replace('<','')
jointure_pop_sn2017['Valeur_y']= pd.to_numeric(jointure_pop_sn2017['Valeur_y'])

#Calcul de la proportion de personne en sous nutrion par rapport au nombre d'habitant et affichage
jointure_pop_sn2017['Montant']= ((jointure_pop_sn2017['Valeur_y']*1000000)/(jointure_pop_sn2017['Valeur_x']*1000))*100
jointure_pop_sn2017.rename(columns={'Valeur_y': 'Nombre de personne en sous nutrition (en million)', 'Valeur_x': "Nombre d'habitant (en millier)"}, inplace=True)
jointure_pop_sn2017 [['Zone',"Nombre d'habitant (en millier)",'Nombre de personne en sous nutrition (en million)','Montant']]
```

```python
#Requete 1 : Proportion de personnes en état de sous nutrition
# Calcul du montant 'Global monde' de personne en sous nutrition
montant_glo_result = (jointure_pop_sn2017['Nombre de personne en sous nutrition (en million)'].sum()*1000000)/(jointure_pop_sn2017["Nombre d'habitant (en millier)"]*1000).sum()
montant_glo_result*100
```

```python
#Requête 2 : Disponibilité alimentaire mondiale
#Agrégation des disponibilités alimentaires pour chaque pays
dispo_ali = dispo_ali.groupby(['Zone']).sum()
#Jointure avec le fichier population
jointure_dispoali_pop = pd.merge(population2017, dispo_ali, on="Zone")
#Calcul de la disponibilité alimentaire par habitant et par année + affichage
jointure_dispoali_pop['Montant_pays'] = (jointure_dispoali_pop['Valeur']*1000)*jointure_dispoali_pop['Disponibilité alimentaire (Kcal/personne/jour)']*365
jointure_dispoali_pop[['Zone','Valeur','Disponibilité alimentaire (Kcal/personne/jour)','Montant_pays']]
```

```python
#Requête 2 : Disponibilité alimentaire mondiale
#Calcul et affichage de la disponibilité alimentaire mondiale
Total_dispo = jointure_dispoali_pop['Montant_pays'].sum()
print ("Disponibilté alimentaire mondiale:", Total_dispo,"kcal")
Total_personne = Total_dispo/(2400*365)
print ("Nombre de personnes pouvant être nourris:",Total_personne/1000000000,"milliards")
print ("Soit : ",((Total_personne/1000000000)/(jointure_dispoali_pop['Valeur'].sum()/1000000))*100,"% de la population mondiale")
```

```python
#Requête 3 : Disponibilité alimentaire mondiale des produits végétaux
#Réimportation du fichier de disponibilité alimentaire
dispo_ali = pd.read_csv('C:/Users/alois/Desktop/Formation Aloïs/Projet 4 - FAO/dispo_alimentaire.csv')
#Isolation des disponibilités alimentaires uniquement d'origine végetale et création d'une dataframe avec les données correspondantes
dispo_ali_veg = dispo_ali['Origine']=='vegetale'
dispo_ali_veg2017 = dispo_ali[dispo_ali_veg]
dispo_ali_veg2017 = dispo_ali_veg2017.groupby('Zone').sum()
#Jointure avec le fichier population et affichage
joint_pop_dispo_ali_veg = pd.merge(population2017, dispo_ali_veg2017, on='Zone')
#Calcul de la disponibilité alimentaire par habitant et par année + affichage
joint_pop_dispo_ali_veg['Montant_pays'] = (joint_pop_dispo_ali_veg['Valeur']*1000)*joint_pop_dispo_ali_veg['Disponibilité alimentaire (Kcal/personne/jour)']*365
joint_pop_dispo_ali_veg[['Zone','Valeur','Disponibilité alimentaire (Kcal/personne/jour)','Montant_pays']]
```

```python
#Requête 3 : Disponibilité alimentaire mondiale des produits végétaux
#Calcul et affichage de la disponibilité alimentaire mondiale d'origine végétale
Total_dispo_veg = joint_pop_dispo_ali_veg['Montant_pays'].sum()
print ("Disponibilté alimentaire mondiale:", Total_dispo,"kcal")
Total_personne_veg = (Total_dispo_veg/(2400*365))
print ("Nombre de personnes pouvant être nourris:",Total_personne_veg/1000000000,"milliards")
print ("Soit : ",((Total_personne_veg/1000000000)/(jointure_dispoali_pop['Valeur'].sum()/1000000))*100,"% de la population mondiale")
```

```python
#Requête 4 : Utilisation de la disponibilité intérieure
#Réimportation du fichier de disponibilité alimentaire + Remplacement des valeurs 'NaN' en 0
dispo_ali = pd.read_csv('C:/Users/alois/Desktop/Formation Aloïs/Projet 4 - FAO/dispo_alimentaire.csv')
dispo_ali = dispo_ali.fillna(0)
#Mise en forme du Dataframe puis agrégation des valeurs
dispo_ali = dispo_ali[['Zone','Produit','Aliments pour animaux','Nourriture','Pertes','Semences','Traitement','Autres Utilisations','Disponibilité intérieure','Production','Importations - Quantité','Exportations - Quantité','Variation de stock']]
dispo_ali_glo = dispo_ali.sum()
#Calcul des différentes proportions par rapport à la disponibilité intérieure + affichage
dispo_ali_glo['Proportion aliments pour animaux'] = (dispo_ali_glo['Aliments pour animaux']*100 )/ dispo_ali_glo['Disponibilité intérieure']
dispo_ali_glo['Proportion Nourriture'] = (dispo_ali_glo['Nourriture']*100 )/ dispo_ali_glo['Disponibilité intérieure']
dispo_ali_glo['Proportion pertes'] = (dispo_ali_glo['Pertes']*100 )/ dispo_ali_glo['Disponibilité intérieure']
dispo_ali_glo
```
